```python
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
from preprocessing import X_scaled, y
```

```python
###Dimensionality Reduction - PCA###

pca= PCA()
X_pca= pca.fit_transform(X_scaled)

# Determine optimal number of components for 90% variance
cumulative_variance = pca.explained_variance_ratio_.cumsum()
n_components_90 = np.argmax(cumulative_variance >= 0.9) + 1
print(f"Number of components to retain 90% variance: {n_components_90}")

# Apply PCA again with optimal number of components
pca_opt = PCA(n_components=n_components_90)
X_pca = pca_opt.fit_transform(X_scaled)

# Plot cumulative explained variance
plt.figure(figsize=(8,5))
plt.plot(cumulative_variance, marker='o')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA - Cumulative Variance")
plt.grid(True)
plt.show()

# Scatter plot using the first 2 components
plt.figure(figsize=(10,6))
plt.scatter(X_pca[:,0],X_pca[:,1], c= y, cmap='viridis')
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.title("PCA - First Two Components")
plt.colorbar(label="Target")
plt.show()
```
